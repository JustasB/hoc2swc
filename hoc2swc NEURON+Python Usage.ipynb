# Imports

```python
from hoc2swc import hoc2swc

import matplotlib.pyplot as plt
from pprint import pprint as pp
```

# Convert a HOC file with a cell to SWC

```python
hoc2swc("tests/testFiles/ca1/CA1.hoc","cell.swc")
```

# Read the SWC file coords

```python
with open('cell.swc', 'r') as f:
    lines = f.readlines()
    
# 3rd-5th columns store XYZ data
x,y,z = zip(*[[float(col) for col in line.split(' ')[2:5]] for line in lines])
```

## Quick plot of the SWC file

```python
plt.scatter(x,z, s=4)
plt.xlim(-400, 400)
plt.show()

plt.scatter(y,z, s=4)
plt.xlim(-400, 400)
plt.show()

plt.scatter(x,y, s=4)
plt.xlim(-400, 400)
plt.show()
```
